```python
import librosa
import librosa.display
import matplotlib.pyplot as plt


(signal, sample_rate) = librosa.load('data/blues/blues.00000.wav')
```

```python
plt.plot(signal)
```

```python
import numpy as np

fft = np.fft.fft(signal)
magnitude = np.abs(fft)
```

```python
frequency = np.linspace(0, sample_rate, len(magnitude))

frequency = frequency[0:int(len(frequency)/2)]
magnitude = magnitude[0:int(len(magnitude)/2)]
```

```python
len(magnitude)
```

```python
n_fft = 2048
hop_length = 512

stft = librosa.stft(signal, n_fft=2048, hop_length=hop_length)
spectrogram = np.abs(stft)
```

```python
spectrogram_db = librosa.amplitude_to_db(stft)
```

```python
librosa.display.specshow(spectrogram_db, sr=sample_rate, hop_length=hop_length, n_fft=n_fft)
plt.colorbar()
```

```python
n_mfcc = 20
mfcc = librosa.feature.mfcc(y=signal, sr=sample_rate, hop_length=hop_length, n_fft=n_fft, n_mfcc=n_mfcc)
```

```python
librosa.display.specshow(mfcc, sr=sample_rate, hop_length=hop_length, n_fft=n_fft)
plt.colorbar()
```
